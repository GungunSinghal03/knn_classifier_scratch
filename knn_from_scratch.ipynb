#KNN

eucledian and manhattan distance

importing libraries

```python
import numpy as np
import statistics
```

**K nearest neighbour**

```python
class KNN_classifier:
  def __init__(self,distance_metrics):
    self.distance_metrics=distance_metrics


  def get_distance_metrics(self,training_data_point,testing_data_point):
    if self.distance_metrics=='eucledian':
      dist=0
      for i in range(len(training_data_point)-1):
        dist=dist+(training_data_point[i]-testing_data_point[i])**2
        eucledian_dist=np.sqrt(dist)
        return eucledian_dist
    elif self.distance_metrics=='manhattan':
      dist=0
      for i in range(len(training_data_point)-1):
        dist=dist+abs(training_data_point-testing_data_point)
        manhattan_dist=dist
        return manhattan_dist



  def get_nearest_neighbour(self,X_train,test_data,k):
    dist_list=[]
    for training_data in X_train:
      distance=self.get_distance_metrics(training_data,test_data)
      dist_list.append(training_data,distance)
      dist_list.sort(key=lambda x:x[1])
      neighbours_list=[]

      for j in range(k):
        neighbours_list.append(dist_list[j][0])
    return neighbours_list


  def predict(self,X_train,test_data,k):
    neighbours=self.get_nearest_neighbour(X_train,test_data,k)
    for data in neighbours:
      label=[]
      label.append(data[-1])
      #-1 means we are only taking target value from the data
    predicted_class=statistics.mode(label)
    return predicted_class



```
